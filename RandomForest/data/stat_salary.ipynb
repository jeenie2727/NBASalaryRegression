```python
import pandas as pd

# Google Drive에서 CSV 파일 경로 지정
salary_final = '/content/drive/MyDrive/ajou/41_ML/project/modified data/salary_final.csv'
new_stat_add5y = '/content/drive/MyDrive/ajou/41_ML/project/modified data/new_stat_add5y.csv'

# salary_deduplication.csv와 2010-2014_stat_aggregated.csv를 로드합니다.
salary_data =  pd.read_csv(salary_final, encoding='ISO-8859-1')
stat_data = pd.read_csv(new_stat_add5y, encoding='ISO-8859-1')
```

```python
# 데이터 타입 일치시키기
salary_data['start'] = salary_data['start'].astype(str)
stat_data['start'] = stat_data['start'].astype(str)

# salary_final에서 필요한 열 선택
salary_data_selected = salary_data[['name', 'start', 'salary', 'target', 'debut']]

# new_stat_add5y에서 필요한 열 선택
stat_data_selected = stat_data[['name', 'start', 'Pos', 'Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']]

# 두 데이터 병합
final_5y = pd.merge(salary_data_selected, stat_data_selected, on=['name', 'start'], how='left')

# 결과를 CSV 파일로 저장
output_file_path = '/content/drive/MyDrive/ajou/41_ML/project/modified data/link_5y.csv'
final_5y.to_csv(output_file_path, index=False)

print(f"File saved to {output_file_path}")
```

```python
# 결측치가 있는 행 찾기
missing_data_indices = final_5y[final_5y.isnull().any(axis=1)].index

for idx in missing_data_indices:
    row = final_5y.loc[idx]
    name = row['name']
    start = row['start']

    # 동일한 name을 가진 데이터 찾기
    similar_rows = final_5y[(final_5y['name'] == name) & (final_5y.index != idx)]

    if not similar_rows.empty:
        # start 값이 현재 행의 start 값보다 작은 행들 중 가장 가까운 행 찾기
        similar_rows['start'] = similar_rows['start'].astype(int)
        start = int(start)
        closest_rows = similar_rows[similar_rows['start'] < start]

        if not closest_rows.empty:
            closest_rows['start_diff'] = abs(closest_rows['start'] - start)
            closest_row_idx = closest_rows['start_diff'].idxmin()
            closest_row = closest_rows.loc[closest_row_idx]

            # 결측치 대체
            for column in final_5y.columns:
                if pd.isnull(row[column]):
                    final_5y.at[idx, column] = closest_row[column]

# Age가 비어 있는 행 삭제
final_5y = final_5y.dropna(subset=['Age'])

# 결과를 CSV 파일로 저장
output_file_path = '/content/drive/MyDrive/ajou/41_ML/project/modified data/final_5y.csv'
final_5y.to_csv(output_file_path, index=False)

print(f"File saved to {output_file_path}")
```
